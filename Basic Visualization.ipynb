# Basic Visualizations

```python

import numpy as np
import matplotlib.pyplot as plt # 2D plotting package
%matplotlib inline 
#to be able to see the visualizations 

# If you dont have pandas or Numpy or Matplotlib download them using pip install or go *here*
```

# 1D Plotting

```python
x = np.linspace(0, 3, 20) # Begins at 0, ends at 3, 20 x coordinates
y = np.linspace (0, 9, 20) # begins at 0, ends at 9, 20 y coordinates
plt.plot(x, y) #line plot
plt.plot(x, y, 'o') # dots on the plot

plt.show # Unecessary when using Jupyter notebooks ;
#(but for other occasions you should know to use this)
```

# 2D Arrays

```python
image = np.random.rand(30, 30) # size of the graph
plt.imshow(image, cmap=plt.cm.gray)
plt.colorbar()
plt.show()
```
